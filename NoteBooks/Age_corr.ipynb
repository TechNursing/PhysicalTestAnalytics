```python
import pandas as pd

Physical_data = pd.read_csv('../Data/Physical_test.csv')
display(Physical_data)
```

```python
# ヒートマップで表すために、カテゴリ型を数値型にget_dymmyで変換する
age_gap = pd.get_dummies(Physical_data, columns=['年代', '体力年齢'], prefix_sep='/')
print(f'age_gap : {age_gap.shape}\n')
display(age_gap.head(5))

# データの形を表示(行数・列数を取得)
print(f'age_gap : {age_gap.shape}\n')

# 各columnのデータ型を表示
print(f'{age_gap.dtypes} \n')
```

```python
# 見やすい様にカラムの順序を変更する
age_gap = age_gap.reindex(columns=['年代/10代男性', '年代/20代男性', '年代/30代男性', '年代/40代男性',
                                   '年代/50代男性', '年代/60代男性', '年代/70代男性', 
                                   '年代/10代女性', '年代/20代女性', '年代/30代女性', '年代/40代女性',
                                   '年代/50代女性', '年代/60代女性', '年代/70代女性', 
                                   '体力年齢/75~79', '体力年齢/70~74', '体力年齢/65~69', '体力年齢/60~64',
                                   '体力年齢/55~59', '体力年齢/50~54', '体力年齢/45~49', '体力年齢/40~44',
                                   '体力年齢/35~39', '体力年齢/30~34', '体力年齢/25~29', '体力年齢/20~24' ])
```

```python
# 年代と体力年齢の相関性を見る
age_gap_corr = age_gap.corr()
age_gap_corr
```

```python
# 見やすい様にカラムの順序を変更する
age_gap_male = age_gap_corr.reindex(columns=['年代/10代男性', '年代/20代男性', '年代/30代男性', '年代/40代男性',
                                             '年代/50代男性', '年代/60代男性', '年代/70代男性'],
                                    index=['体力年齢/75~79', '体力年齢/70~74', '体力年齢/65~69', '体力年齢/60~64',
                                           '体力年齢/55~59', '体力年齢/50~54', '体力年齢/45~49', '体力年齢/40~44',
                                           '体力年齢/35~39', '体力年齢/30~34', '体力年齢/25~29', '体力年齢/20~24'])
```

```python
# 見やすい様にカラムの順序を変更する
age_gap_female = age_gap_corr.reindex(columns=['年代/10代女性', '年代/20代女性', '年代/30代女性',
                                             '年代/40代女性', '年代/50代女性'],
                                    index=['体力年齢/75~79', '体力年齢/70~74', '体力年齢/65~69', '体力年齢/60~64',
                                           '体力年齢/55~59', '体力年齢/50~54', '体力年齢/45~49', '体力年齢/40~44',
                                           '体力年齢/35~39', '体力年齢/30~34', '体力年齢/25~29', '体力年齢/20~24'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib

# 文字化けするので、フォントを指定
sns.set_theme(font_scale=1.5, font='IPAexGothic')
```

```python
# カラム量が多いため、サイズを見やすく調整
fig, ax = plt.subplots(figsize=(6, 6)) 

# ヒートマップで可視化
plt.tight_layout()
sns.heatmap(age_gap_male, cmap="Blues", square=True).set(title='年代と体力年齢の相関関係（男）')

results_path = '../Graphs/年代と体力年齢の相関関係（男）.png'
plt.savefig(results_path, bbox_inches='tight')
```

```python
# カラム量が多いため、サイズを見やすく調整
fig, ax = plt.subplots(figsize=(6, 6)) 

# ヒートマップで可視化
plt.tight_layout()
sns.heatmap(age_gap_female, cmap="Blues", square=True).set(title='年代と体力年齢の相関関係（女）')

results_path = '../Graphs/年代と体力年齢の相関関係（女）.png'
plt.savefig(results_path, bbox_inches='tight')
```
