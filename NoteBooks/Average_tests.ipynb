```python
import pandas as pd

physical_data = pd.read_csv('../Data/Physical_test.csv')
# display(physical_data)
```

```python
# 年代カラムの値を性別分類無しの、年代のみに置換
physical_data['年代'] = physical_data['年代'].str.replace('.性','', regex=True)
gene_list = physical_data["年代"].unique()
print("一意の「close_contact」リスト\n", gene_list)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib

# 文字化けするので、フォントを指定
sns.set_theme(font_scale=1.5, font='IPAexGothic')
```

```python
# x軸を見やすいように順番入れ替える
order=['10代', '20代', '30代', '40代', '50代', '60代', '70代']
physical_data['年代'] = pd.Categorical(physical_data['年代'], order)
```

```python
# グラフを見やすいように背景色を変更する
plt.figure(facecolor='#C1CDCD')
ax = plt.axes()
ax.set_facecolor('#F0FFFF')

# 分布図をseabornのscatterplotで可視化
sns.scatterplot(data=physical_data, x="年代", y="握力平均", hue="性別").set(title='握力の年代・性別分布と全国平均')

# 全国平均の折れ線グラフを足す
import numpy as np

man = np.array([0, 1, 2, 3, 4, 5, 6])
m_height = np.array([43, 46.5, 47, 46.7, 45, 41, 30])

woman = np.array([0, 1, 2, 3, 4, 5, 6])
w_height = np.array([27, 27.5, 29, 29.3, 27.8, 25.8, 23])

# sns.scatterplotの上に折れ線グラフを置く
plt.plot(man, m_height, label="全国平均（男）")
plt.plot(woman, w_height, label="全国平均（女）")

# 凡例の位置をグラフと重複しないように修正する
plt.legend(loc='upper right', bbox_to_anchor=(1.55, 1))

# グラフを画像保存する
results_path = '../Graphs/握力の年代・性別分布と全国平均.png'
plt.savefig(results_path, bbox_inches='tight')
```
