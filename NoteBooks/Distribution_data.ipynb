```python
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib

sns.set_theme(font_scale=1.5, font='IPAexGothic')
```

```python
# 全体の測定の分布
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center', va = 'bottom', c = 'Magenta')


plt.figure(figsize=(16, 7))
x=['握力（右）', '握力（左）', '握力平均', '上体起こし', '上体そらし', '背筋力',
   '立ち上がり', '30秒立ち座り', '垂直跳び', '2ステップ', '2ステップの距離',
   '立ち幅跳び', '反復横跳び', '20mシャトルラン', 'ハンドボール投げ',
   '長座体前屈', '閉眼片足立ち', '棒反応', '肺活量']
height=[
    Physical_data["握力（右）"].count(),Physical_data["握力（左）"].count(),
    Physical_data["握力平均"].count(),Physical_data["上体起こし"].count(),
    Physical_data["上体そらし"].count(),Physical_data["背筋力"].count(),
    Physical_data["立ち上がり"].count(),Physical_data["30秒立ち座り"].count(),
    Physical_data["垂直跳び"].count(),Physical_data["2ステップ"].count(),
    Physical_data["2ステップの距離"].count(),Physical_data["立ち幅跳び"].count(),
    Physical_data["反復横跳び"].count(),Physical_data["20mシャトルラン"].count(),
    Physical_data["ハンドボール投げ"].count(),Physical_data["長座体前屈"].count(),
    Physical_data["閉眼片足立ち"].count(),Physical_data["棒反応"].count(),
    Physical_data["肺活量"].count()]
plt.bar(x=x, height=height)

addlabels(x, height)
plt.xticks(rotation=60)
plt.title('総種目数の各実施数', fontsize=18)
plt.box(True)
results_path = '../Graphs/総種目数の各実施数.png'
plt.savefig(results_path, bbox_inches='tight')
plt.show()
```
