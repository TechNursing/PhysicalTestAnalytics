```python
# 必要なカラムのみ抽出
qual_physi_data = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '握力（右）', '握力（左）', '握力平均',
                                          '上体起こし', '上体そらし', '背筋力', '立ち上がり', '30秒立ち座り', '垂直跳び',
                                          '2ステップ', '2ステップの距離', '立ち幅跳び', '反復横跳び', '20mシャトルラン',
                                          'ハンドボール投げ', '長座体前屈', '閉眼片足立ち', '棒反応', '肺活量']]
display(qual_physi_data)
```

```python
# それぞれの測定項目を１つのカラム内に納める
grip_r = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '握力（右）']]
grip_l = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '握力（左）']]
grip_m = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '握力平均']]
crunch = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '上体起こし']]
backward = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '上体そらし']]
spine = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '背筋力']]
stand_up = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '立ち上がり']]
seconds_30 = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '30秒立ち座り']]
vertical_jump = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '垂直跳び']]
steps_2 = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '2ステップ']]
steps_2_dis = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '2ステップの距離']]
stand_jump = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '立ち幅跳び']]
side_jump = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '反復横跳び']]
shuttle_run = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '20mシャトルラン']]
handball = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', 'ハンドボール投げ']]
sit_bend = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '長座体前屈']]
stand_closed = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '閉眼片足立ち']]
stick_react = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '棒反応']]
capacity = qual_physi_data.loc[:, ['生まれ年', '年代', '体力年齢', '性別', '身長', '肺活量']]
display(spine)
```

```python
# カラム名を統一し、mergeする
grip_r = grip_r.rename(columns = {"握力（右）": "tests"})
grip_l = grip_l.rename(columns = {"握力（左）": "tests"})
grip_m = grip_m.rename(columns = {"握力平均": "tests"})
crunch = crunch.rename(columns = {"上体起こし": "tests"})
backward = backward.rename(columns = {"上体そらし": "tests"})
spine = spine.rename(columns={"背筋力": "tests"})
stand_up = stand_up.rename(columns = {"立ち上がり": "tests"})
seconds_30 = seconds_30.rename(columns = {"30秒立ち座り": "tests"})
vertical_jump = vertical_jump.rename(columns = {"垂直跳び": "tests"})
steps_2 = steps_2.rename(columns={"2ステップ": "tests"})
steps_2_dis = steps_2_dis.rename(columns = {"2ステップの距離": "tests"})
stand_jump = stand_jump.rename(columns = {"立ち幅跳び": "tests"})
side_jump = side_jump.rename(columns = {"反復横跳び": "tests"})
shuttle_run = shuttle_run.rename(columns={"20mシャトルラン": "tests"})
handball = handball.rename(columns = {"ハンドボール投げ": "tests"})
sit_bend = sit_bend.rename(columns = {"長座体前屈": "tests"})
stand_closed = stand_closed.rename(columns = {"閉眼片足立ち": "tests"})
stick_react = stick_react.rename(columns={"棒反応": "tests"})
capacity = capacity.rename(columns={"肺活量": "tests"})
display(spine)
```

```python
# mergeする
grip_r_l = pd.merge(grip_r, grip_l, how = 'outer')
grip_m_crunch = pd.merge(grip_m, crunch, how = 'outer')
backward_spine = pd.merge(backward, spine, how = 'outer')
stand_up_seconds_30 = pd.merge(stand_up, seconds_30, how = 'outer')
vertical_jump_steps_2 = pd.merge(vertical_jump, steps_2, how = 'outer')
steps_2_dis_stand_jump = pd.merge(steps_2_dis, stand_jump, how = 'outer')
side_jump_shuttle_run = pd.merge(side_jump, shuttle_run, how = 'outer')
handball_sit_bend = pd.merge(handball, sit_bend, how = 'outer')
stand_closed_stick_react = pd.merge(stand_closed, stick_react, how = 'outer')
stand_closed_capacity = pd.merge(stand_closed_stick_react, capacity, how = 'outer')
display(backward_spine)
```

```python
grip_crunch = pd.merge(grip_r_l, grip_m_crunch, how = 'outer')
backward_seconds_30 = pd.merge(backward_spine, stand_up_seconds_30, how = 'outer')
vertical_jump_stand_jump = pd.merge(vertical_jump_steps_2, steps_2_dis_stand_jump, how = 'outer')
side_jump_sit_bend = pd.merge(side_jump_shuttle_run, handball_sit_bend, how = 'outer')
```

```python
grip_seconds_30 = pd.merge(grip_crunch, backward_seconds_30, how = 'outer')
vertical_jump_sit_bend = pd.merge(vertical_jump_stand_jump, side_jump_sit_bend, how = 'outer')
```

```python
grip_sit_bend = pd.merge(grip_seconds_30, vertical_jump_sit_bend, how = 'outer')
all_data = pd.merge(grip_sit_bend, stand_closed_capacity, how = 'outer')
display(all_data)
tests_list = all_data["tests"].unique()
print("一意の「tests」リスト\n", tests_list)
```

```python
# 値が存在するもののみ抽出する
all_data = all_data[all_data['tests'] != "False"]
display(all_data)
tests_list = all_data["tests"].unique()
print("一意の「tests」リスト\n", tests_list)
```
